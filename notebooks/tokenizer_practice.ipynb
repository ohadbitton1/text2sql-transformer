```python

import pandas as pd
df = pd.read_csv('../data/validation.csv')
print(df.head())



question_from_data = df['question'][3]
t5_prefix = "translate English to SQL:"
full_request = t5_prefix + question_from_data
print(full_request)
```

```python
from transformers import AutoTokenizer
#load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("t5-small")
input_ids = tokenizer(full_request)
lables = tokenizer(df['sql'][3])
print("--input ids---")
print(input_ids)
print("----lables-------")
print( lables )
```

```python
decoded_text = tokenizer.decode(input_ids ['input_ids'])
decoded_text2 = tokenizer.decode(lables['input_ids'])
print(decoded_text)
print(decoded_text2)
```

```python
# full tokenization
q_list = df['question']
sql_list = df['sql'].tolist()
full_request2 = (t5_prefix + q_list).tolist()

# הפעלת הטוקנייזר על כל הרשימה, זה בסדר
input_ids_full = tokenizer(full_request2)
lables_full = tokenizer(sql_list)

# 1. יצירת רשימה של אורכי השאלות
question_lengths = [len(item) for item in input_ids_full['input_ids']] 
# הפריט כאן הוא רשימת הטוקנים של שאלה בודדת.

# 2. יצירת רשימה של אורכי ה-SQL
sql_lengths = [len(item) for item in lables_full['input_ids']]
# אתה משלים את החסר!

# 3. מציאת האורך המקסימלי
max_q_length = max(question_lengths)
max_sql_length = max(sql_lengths)

print(f"Maximum Question Length (tokens): {max_q_length}")
print(f"Maximum SQL Length (tokens): {max_sql_length}")

# שים לב: מחקנו את שורות ה-full_decoded שגרמו לשגיאה!
```
